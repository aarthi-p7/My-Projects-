```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('customers.csv')
print(df.head())
print("\nðŸ“Š Data Summary:")
print(df.describe())
plt.figure(figsize=(5, 4))
sns.countplot(data=df, x='Gender', palette='pastel')
plt.title('Customer Gender Distribution')
plt.tight_layout()
plt.show()
plt.figure(figsize=(6, 4))
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.tight_layout()
plt.show()
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='AnnualIncome', y='SpendingScore', hue='Gender', palette='Set2')
plt.title('Income vs Spending Score')
plt.tight_layout()
plt.show()
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='Gender', y='SpendingScore', palette='cool')
plt.title('Spending Score by Gender')
plt.tight_layout()
plt.show()
plt.figure(figsize=(5, 4))
sns.heatmap(df[['Age', 'AnnualIncome', 'SpendingScore']].corr(), annot=True, cmap='Blues')
plt.title('Correlation Between Features')
plt.tight_layout()
plt.show()
```
