```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'Month': ['January', 'February', 'March', 'April', 'May'],
    'Product A': [1200, 1350, 1600, 1400, 1550],
    'Product B': [950, 1050, 1200, 1250, 1300],
    'Product C': [600, 700, 800, 900, 950]
}
df = pd.DataFrame(data)
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
for product in ['Product A', 'Product B', 'Product C']:
    plt.plot(df['Month'], df[product], marker='o', label=product)
plt.title(' Monthly Sales Trend by Product')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
total_sales = df[['Product A', 'Product B', 'Product C']].sum().reset_index()
total_sales.columns = ['Product', 'Total Sales']
plt.figure(figsize=(7, 5))
sns.barplot(x='Product', y='Total Sales', data=total_sales, palette='Set2')
plt.title(' Total Sales per Product (JanMay)')
plt.ylabel('Total Sales')
plt.xlabel('Product')
plt.tight_layout()
plt.show()
growth_df = df.copy()
for product in ['Product A', 'Product B', 'Product C']:
    growth_df[product] = growth_df[product].pct_change().round(3) * 100
growth_df = growth_df.fillna(0)
growth_df.set_index('Month', inplace=True)
print(" Month-over-Month Growth Rate (%):")
display(growth_df.style.format("{:.1f}%").set_caption("Monthly Growth (%) for Each Product"))
```
